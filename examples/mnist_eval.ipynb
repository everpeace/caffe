# LeNet(畳み込みニューラルネット)によるMNIST(hand-written digits)データの認識

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

caffe_root = '../'  # this file is expected to be in {caffe_root}/examples
import sys
sys.path.insert(0, caffe_root + 'distribute/python')

import caffe
from utils import tile_raster_images
```

```python
# LeNet読み込み
MODEL_FILE = './mnist/lenet.prototxt'
PRETRAINED = './mnist/lenet_iter_10000.caffemodel'
net = caffe.Classifier(MODEL_FILE, PRETRAINED)
net.set_phase_test()
net.set_mode_gpu()
```

```python
# 実験データ読み込み
from load_mnist_data import load_data
train_set, valid_set, test_set = load_data('mnist.pkl.gz')
(train_set_x, train_set_y) = train_set
(valid_set_x, valid_set_y) = valid_set
(test_set_x , test_set_y) = test_set
```

## LeNetの構造

```python
from google.protobuf import text_format
from caffe.draw import get_pydot_graph
from caffe.proto import caffe_pb2
from IPython.display import display, Image 

_net = caffe_pb2.NetParameter()
f = open(MODEL_FILE)
text_format.Merge(f.read(), _net)
display(Image(get_pydot_graph(_net).create_png()))
```

## テストデータセットでの予測

```python
test_set_x = test_set_x.reshape((test_set_x.shape[0], 28, 28, 1))
import copy
predictions_raw = net.predict(test_set_x, oversample=False)
predictions = np.apply_along_axis(np.argmax, 1, predictions_raw)
wrong_idx = np.where(predictions!=test_set_y)[0]
```

### 予測例

```python
import matplotlib.cm as cm
idx = 1024
plt.imshow(test_set_x[idx].reshape((28,28)), cmap = cm.Greys_r)
print '予測結果 :  %d    (答: %d)' % (predictions[idx], test_set_y[idx])
```

## 予測精度

```python
print '\n予測精度: %2.3f ％   (%d件中 %d 件正解)' % (  (1- len(wrong_idx)/float(len(test_set_y))) *100, 
                                                                                   test_set_y.shape[0], 
                                                                                   test_set_y.shape[0]-len(wrong_idx) )
plt.figure(figsize=(10,10))
plt.imshow(tile_raster_images(X=test_set_x[wrong_idx], img_shape=(28, 28), tile_shape=(wrong_idx.shape[0]/20 + 1,20), tile_spacing=(1, 1)), 
                    cmap=cm.Greys_r)
print '\n予測エラー(予測/正解)\n'
a = test_set_y[wrong_idx]
wrong_p = predictions[wrong_idx]
sys.stdout.write('       ')
for i in xrange(wrong_idx.shape[0]):
    import sys
    sys.stdout.write('%d/%d   ' % (wrong_p[i],a[i]) )
    if (i % 20) ==19 : sys.stdout.write('\n       ')
print "\n\n予測エラーのテストイメージ"
```
