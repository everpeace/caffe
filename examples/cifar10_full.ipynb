# 畳み込みニューラルネットワーク(Alex Krizhevsky's CNN)によるCIFAR-10データでの物体認識
この例はAlex Krizhevsky氏の[cuda-convnet](https://code.google.com/p/cuda-convnet/)による畳み込みニューラルネットを用いたCIFAR-10データでの物体認識デモをCaffeで再現したものです。PyCaffe(CaffeのPython インターフェース)のデモにもなっています。

## ニューラルネットを読み込む(再構成)
まず学習済みのニューラルネットを読み込みます。モデルファイルと、その重みファイルからClassifierを生成します。

```python
MODEL_FILE = './cifar10/cifar10_full.prototxt'
PRETRAINED_WEIGHTS = './cifar10/cifar10_full_iter_70000.caffemodel'

# pytyhon pathにpycaffeを追加(念のため)
import sys
caffe_root = '../'  # this file is expected to be in {caffe_root}/examples
sys.path.insert(0, caffe_root + 'distribute/python')

# CNNをモデルファイル&重みファイルから再構成
import caffe
from caffe.proto import caffe_pb2
from caffe.io import blobproto_to_array
net = caffe.Classifier(MODEL_FILE, PRETRAINED_WEIGHTS)
net.set_mode_gpu() #GPUモード設定
```

学習時にWhitening(イメージの平均値を0にする処理)をしたので、利用時も入力からその平均値(mean)を引いてやる必要があります。入力のmeanはdata層の名前を指定して、set_meanメソッドで設定します。

_注:通常であれば、後述のpredictメソッドで自動で処理されるはずですが、うまく効かないため、手動でmeanを引くことにしたのでここではセットしないことにします_

```python
MEAN_FILE = "./cifar10/mean.binaryproto"
blob = caffe_pb2.BlobProto()
blob.ParseFromString(open(MEAN_FILE, "rb").read())
mean = blobproto_to_array(blob).reshape(3, 32, 32)
# net.set_mean('data', mean)  # 'data' レイヤにmeanを設定
```

この後使う色々な部品を個々で一気に読み込んでいます。(読み飛ばしてください)

```python
# その他 色々必要なライブラリをインポート
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline
import leveldb

# テストデータセット読み込み
db = leveldb.LevelDB('./cifar10/cifar10_test_leveldb/')
test_data_raw =  [v for v in db.RangeIter()]

# Rawデータからprotocolbufferのdatumに変換
def toDatum(x):
    return caffe_pb2.Datum.FromString(x[1])
test_datum_set = map(toDatum, test_data_raw)

# datumからデータ(イメージ)とラベルを抽出
# イメージデータの次元(shape)は(N, C(色数), Height ,Width)なので、(N, Height, Width, C)に変換しておく
test_data = np.rollaxis(np.array(map(caffe.io.datum_to_array, test_datum_set)),1,4)
test_label = np.array(map(lambda x: x.label, test_datum_set))

# ラベル名 (数値 => 物体名) 
# (python向けのcifar-10データ付属のmetaデータから読み込み)
import cPickle
fo = open('cifar10/cifar-10-batches-py/batches.meta', 'rb')
meta = cPickle.load(fo)
fo.close()
label_names = np.array(meta['label_names'])

# イメージプロット用ユーティリティ(複数のイメージをタイル上(tile_shapeで指定)にプロット)
#  imagesの次元(shape)は、RGBの場合=>(N, H, W, 3) or 白黒の場合=>(N, H, W)
def plot_images(images, tile_shape):
    assert images.shape[0] <= (tile_shape[0]* tile_shape[1])
    from mpl_toolkits.axes_grid1 import ImageGrid
    fig = plt.figure()
    grid = ImageGrid(fig, 111,  nrows_ncols = tile_shape ) 
    for i in range(images.shape[0]):
        grd = grid[i]
        grd.imshow(images[i])
```

## このニューラルネットワークの構造
PyCaffeにはdrawパッケージが付属していて、ニューラルネットの構造を[dot format](http://www.graphviz.org/content/dot-language)で出力出来ます。畳み込み層->プーリング層(ReLU適用)->正規化層が3つ連なっていて、最後に10出力ニューロンに全接続でつなぎこまれているのが解ります。最後にsoftmax関数で確率(確信度)として出力するようになっています。

```python
from google.protobuf import text_format
from caffe.draw import get_pydot_graph
from IPython.display import display, Image 

net_param = caffe_pb2.NetParameter()
f = open(MODEL_FILE)
text_format.Merge(f.read(), net_param)
display(Image(get_pydot_graph(net_param).create_png()))
```

## 実際にニューラルネットに予測させてみる
では、読み込んだニューラルネットにテストデータを入力して予測させてみましょう。

まずは、テストデータ(物体名と実際の画像)を確認してみましょう。各画像に物体名が付与されています。

```python
print [label_names[i] for i in test_label[0:32]]
plot_images(test_data[0:32], tile_shape=(4,8))
```

### 実際にテストデータを入力して予測してみる
テストデータ(画像)をニューラルネットに入力して予測させてみましょう。予測させるには`predict`メソッドを使います。

_注: 通常はmeanが自動で処理されるはずですが、なぜかうまく効かないので手動でmeanを引いています。_

```python
input_data = [d - mean.transpose(1, 2, 0) for d in test_data]
predictions_raw = net.predict(input_data)
```

このニューラルネットの出力次元は

```python
print "predictions_raw.shape = ", predictions_raw.shape, "  #10000は入力件数"
print "predictions_raw[0]       = ", ["%3.3f" % p for p in predictions_raw[0]]
```

となっており、各物体の確信度が出力されるようになっています。

では、早速予測結果を見てみましょう。その前に、ニューラルネットの出力は各物体の確信度になっているので、`np.argmax`を利用して、一番確信度の高い物体を予測結果として採用することにします。

```python
predictions = np.apply_along_axis(np.argmax, 1, predictions_raw)
plot_images(test_data[0:32], tile_shape=(4,8))
print label_names[predictions[0:32]]
```

概ね正しく予測できていることが確認できると思います。

### エラー率
次に、予測結果全体のエラー率を見てみましょう。全体でこのくらいの誤検出率であることが解ります。

```python
assert len(predictions)==len(test_label)
error_idx = np.where(predictions!=test_label)[0]
print "エラー率: %3.2f percent" % ( len(error_idx)/float(len(predictions))*100)
print "               (%d/%d images)" % (len(error_idx), len(test_label))
```

### 誤検出されたデータ
実際に誤検出されたデータを見てみましょう。

```python
idx = error_idx[900:920]
input_image = test_data[idx]
wrong_answers = [(label_names[v[0]], label_names[v[1]]) for v in zip(test_label[idx], predictions[idx])]
print ["(ANS: %s, PRED: %s)"% (v[0], v[1]) for v in wrong_answers]
plot_images(input_image,tile_shape=(2,10))
```

例えば、このイメージを見てみましょう。

```python
target = error_idx[900]
plt.imshow(test_data[target])
print "      (ans: %s, prediction: %s)" % (label_names[test_label[target]], label_names[predictions[target]])
```

このイメージの正答は言うまでも無く「馬」ですが、このニューラルネットは、「犬」だと答えてしまっています。(犬に見えないこともないかもしれません。)

実際のニューラルネットの出力である確信度を見てみましょう。

```python
p = predictions_raw[target]
plt.yticks(np.array(range(len(p)))+0.5, tuple(label_names))
plt.barh(range(len(p)), p, height =0.8)
```

かなりの確信度で犬だと誤解しているようですが、馬かもしれない程度には認識出来ているのが確認できます。(これは学習用データへの過学習が原因だと考えられます。パラメータをチューニングしたり、学習用データにもっと馬の画像を多く含めたりすることで改善できる可能性があります。)

## 各層の出力を可視化する
ここでは、ニューラルネットの中をどのように画像が変換されながら伝わっているのかを可視化してみましょう。

各層の次元はこのようになっています。

```python
net.predict([input_data[target]])
blobs = net.blobs
[(k, v.data[0].shape) for k, v in blobs.items()]
```

最初のdata層はRGBの3チャネル, 32x32イメージ入力であることを示しており、それ以降のconv, pool層の次元は、(フィルタ数, height, width)であることを示しており、ip1, probはニューロン数がそれぞれ10であることを示しています。

下記に、conv1層から順に、馬の画像を入力した時の各層の出力を示します。

まずは'conv1'です。conv1は畳み込み層で、RGBの32x32の画像に32個のフィルターをかける処理をする層です(RGBの3層は1次元に畳み込まれます)。

```python
plot_images(np.array(map(np.fliplr , blobs['conv1'].data[0])), tile_shape=(4,8))
```

pool1は、conv1の出力をダウンサンプリングする層です(正確には各downsample用のwindowをスライドさせながら、windoew内のmaxを取り、ReLU関数を通しています)。学習されたニューロン群は、輪郭を抽出しているような処理を行うようになっているのが見て取れます。

```python
plot_images(np.array(map(np.fliplr , blobs['pool1'].data[0])), tile_shape=(4,8))
```

conv2, pool2, conv3, pool3層とどんどん深く潜って行きましょう。conv2, pool2のニューラルネットの構造はconv1, pool1と相似形です(画像のサイズが16x16で、ダウンサンプルじに平均値を取っている点が異なる)。

```python
plot_images(np.array(map(np.fliplr , blobs['conv2'].data[0])), tile_shape=(4,8))
```

```python
plot_images(np.array(map(np.fliplr , blobs['pool2'].data[0])), tile_shape=(4,8))
```

conv3, pool3はconv2, pool2と相似形でフィルタ数が64個で、画像サイズが8x8になっています。

```python
plot_images(np.array(map(np.fliplr , blobs['conv3'].data[0])), tile_shape=(4,16))
```

```python
plot_images(np.array(map(np.fliplr , blobs['pool3'].data[0])), tile_shape=(4,16))
```

この辺りまで来ると、各画像が、入力画像のどの性質を表しているかだんだんわかりにくくなってきますが、これらの小さな画像断片が物体認識のための特徴量になっています。

ip1層はpool3の出力層(正確にはpool3の後にnorm3という正規化層がありますが詳細は割愛します)を入力として、10個の出力ニューロンへつながっています。例えばip層の0番目のニューロンは「飛行機」用のニューロンで、「飛行機」画像の場合によく反応するように、pool3の出力画像ごとの重みを学習しています。

ip1層の出力はこのようになっています。

```python
plt.yticks(np.array(range(10))+0.5, tuple(label_names))
plt.barh(range(10), blobs['ip1'].data[0].reshape(10), height =0.8)
```

prob層はsoftmaxという関数を使って、ip1層の出力を指数関数を使って差を増幅し、総和が1になるように正規化します。

```python
plt.yticks(np.array(range(10))+0.5, tuple(label_names))
plt.barh(range(10), blobs['prob'].data[0].reshape(10), height =0.8)
```

## 参考: 各層のフィルタを可視化する
この節では、各層のフィルタを可視化してみます。今回の畳み込みニューラルネットワークで利用されたフィルタはすべて5x5のフィルタです。この5x5のフィルタを各層の入力画像に1ピクセルずつスライドさせながら適用して、その層の出力を得ます(正確には、出力画像を入力画像と同サイズにするために、入力画像の外側に2ピクセルのpaddingを行っています)。

各フィルタ(重み)の次元は次のとおりです。

```python
# dimentions of filters
[(k,(v[0].data.shape, v[1].data.shape)) for k, v in net.params.items()]
```

conv1層のフィルタは入力data層のRGBの3チャネルごとに5x5サイズの32個をフィルタがあります。

```python
f = net.params['conv1'][0].data
f = np.rollaxis(f, 1,4)
plot_images(np.array(map(np.fliplr, f)), tile_shape=(4,8))
```

conv2層は、conv1層の各フィルタ32個ごとに32個のフィルタがあります。(conv1層の出力はすべて畳み込み計算が行われるため、conv2層の32個のフィルタはconv1層向けの32個のフィルタをすべて足しあわせて可視化しています。)

```python
f = np.array(map(sum, net.params['conv2'][0].data))
plot_images(np.array(map(np.fliplr, f)), tile_shape=(4,8))
```

conv3はconv2と同じように、前層のフィルタ数32ごとに64個のフィルタが存在しているので、それぞれの32個を足しあわせたフィルタ64個を可視化しています。

```python
f = np.array(map(sum, net.params['conv3'][0].data))
plot_images(np.array(map(np.fliplr, f)), tile_shape=(4,16))
```

ip1層はpool3の出力である各画像の各ピクセルごとに(総数は`(64*4*4=1024)`)がそれぞれ10個のニューロンにつながっています。下記は「飛行機」ニューロンへつなぎこまれている重みを可視化したものです。

```python
obj_idx = 1
f = net.params['ip1'][0].data.reshape(10,1024).reshape(10,64,4,4)
print "ip1層における", label_names[obj_idx], "用ニューロンへのpool3層の出力画像毎の重み"
plot_images(np.array(map(np.fliplr, f[obj_idx])), tile_shape=(4,16))
```
