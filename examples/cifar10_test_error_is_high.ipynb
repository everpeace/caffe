# In manual test, test error of 'cifar10_quick' is about 91%.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline

import leveldb

import sys
caffe_root = '../'  # this file is expected to be in {caffe_root}/examples
sys.path.insert(0, caffe_root + 'distribute/python')

import caffe
from caffe.proto import caffe_pb2

# CNN reconstruction and load trained weights.
MODEL_FILE = './cifar10/cifar10_quick.prototxt'
PRETRAINED = './cifar10/cifar10_quick_iter_5000.caffemodel'
net = caffe.Classifier(MODEL_FILE, PRETRAINED)
net.set_phase_test()
net.set_mode_gpu()

# utility function: plot multiple images arranging in tiles.
def plot_images(images, tile_shape, cmap = cm.Greys_r):
    assert images.shape[0] <= (tile_shape[0]* tile_shape[1])
    from mpl_toolkits.axes_grid1 import ImageGrid
    fig = plt.figure()
    grid = ImageGrid(fig, 111,  nrows_ncols = tile_shape ) 
    for i in range(images.shape[0]):
        grd = grid[i]
        grd.imshow(images[i], cmap=cmap)
```

```python
# Load test data set
db = leveldb.LevelDB('./cifar10/cifar10_test_leveldb/')
test_data_raw =  [v for v in db.RangeIter()]

# convert raw data to datum
def toDatum(x):
    return caffe_pb2.Datum.FromString(x[1])
test_datum_set = map(toDatum, test_data_raw)

# extract datum to image data and label
test_data = np.array(map(caffe.io.datum_to_array, test_datum_set))
test_label = np.array(map(lambda x: x.label, test_datum_set))

# Load label names, borrowed from cifar-10 dataset for python.
import cPickle
fo = open('cifar10/cifar-10-batches-py/batches.meta', 'rb')
meta = cPickle.load(fo)
fo.close()
label_names = meta['label_names']
```

## Check loaded test data and lables (you can see it's correct or not by your eyes)

```python
print [label_names[i] for i in test_label[0:20]]
plot_images(np.rollaxis(test_data[0:20],1,4), tile_shape=(4,5))
```

## Test error is 90.88%.... What went wrong??

```python
# convert image data dimension (#num, 3, 32, 32) => (#num, 32, 32, 3)
predictions_raw = net.predict(np.rollaxis(test_data,1,4))
```

```python
predictions = np.apply_along_axis(np.argmax, 1, predictions_raw)
error_idx = np.where(predictions!=test_label)[0]
print "Test error: %3.2f percent" % ( len(error_idx)/float(len(predictions))*100)
```
